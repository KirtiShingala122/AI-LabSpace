```python
# importing the basic libraries
import nltk
import networkx as nx
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
import string
```

```python
# downloading the packages
nltk.download('punkt')
nltk.download('stopwords')
```

```python
text="""Artificial intelligence is transforming modern industries through automation and data-driven insights.
Machine learning algorithms help systems learn patterns and make accurate predictions.
Natural language processing enables computers to understand and generate human language. """
```

```python
def textrank_keywords(text ,topn=5,window_size=4):
  word =word_tokenize(text.lower())
  stop_words = set(stopwords.words('english'))
  words=[w for w in word if w not in stop_words and w not in string.punctuation]
  graph=nx.Graph()
  for i,word in enumerate(words):
    for j in range(i+1,i+window_size):
      if(j<len(words)):
        graph.add_edge(word,words[i])
  scores=nx.pagerank(graph)
  ranked_words=sorted(scores.items(),key=lambda x:x[1],reverse=True)
  keywords=[word for word, score in ranked_words[:topn]]
  return keywords
```

```python
t1="""India is a diverse country located in South Asia with a rich cultural heritage.
It is the seventh-largest country in the world by land area and the most populous country.
India is known for its unity in diversity, with many languages, religions, and traditions.
The capital of India is New Delhi, and it is the largest democracy in the world.
India has made significant advancements in technology, space research, and education.  """
```

```python
nltk.download('punkt_tab')
```

```python
keywords =textrank_keywords(t1,topn=8)
print(keywords)
```

Ngram based next word prediction

```python
import nltk
from nltk.util import ngrams
from collections import Counter, defaultdict
import random
```

```python
nltk.download('punkt')
```

```python
corpus = """
I like machine learning
I like deep learning
I enjoy machine learning
I enjoy data science
Machine learning is interesting
Deep learning is powerful
"""
```

```python
tokens=nltk.word_tokenize(corpus.lower())
```

```python
# create n-gram models
n=2
bigrams =list(ngrams(tokens,n))
```

```python
bigrams
```

```python
# creating the probability table\
model = defaultdict(Counter)
for w1,w2 in bigrams:
  model[w1][w2]+=1
```

```python
model
```

```python
def predict_nextword(word):
  if word in model:
    next_words=model[word]
    prediction=max(next_words,key=next_words.get)
    return prediction
  else:
    return "No prediction "
```

```python
input_word="I"
print(predict_nextword(input_word.lower()))
```
